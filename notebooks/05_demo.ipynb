# 05 - Demo

This notebook checks the demo setup with normal Python imports, then launches the existing Streamlit app as a subprocess.

## Step 1 - Setup

```python
from pathlib import Path
import subprocess
import sys
import time

def find_notebooks_dir():
    cwd = Path.cwd().resolve()
    candidates = [cwd, cwd / "notebooks", cwd / "salient-object-detection-cnn" / "notebooks"]
    for parent in cwd.parents:
        candidates.extend([parent / "notebooks", parent / "salient-object-detection-cnn" / "notebooks"])
    for candidate in candidates:
        if (candidate / "notebook_utils.py").exists():
            return candidate
    raise FileNotFoundError("Could not find notebooks/notebook_utils.py")

NOTEBOOKS_DIR = find_notebooks_dir()
sys.path.insert(0, str(NOTEBOOKS_DIR))

from notebook_utils import PROJECT_DIR, PYTHON

sys.path.insert(0, str(PROJECT_DIR))
print(f"Using project directory: {PROJECT_DIR}")
```

## Step 2 - Verify Demo Checkpoint

```python
checkpoint = PROJECT_DIR / "checkpoints/best_model.pth"
typed_baseline = PROJECT_DIR / "checkpoints/best_model_baseline.pth"
typed_unet = PROJECT_DIR / "checkpoints/best_model_unet_small.pth"

print("default demo checkpoint:", checkpoint, checkpoint.exists())
print("baseline checkpoint:", typed_baseline, typed_baseline.exists())
print("unet_small checkpoint:", typed_unet, typed_unet.exists())

if not checkpoint.exists():
    print("Run a training notebook first. demo_app.py expects checkpoints/best_model.pth.")
```

## Step 3 - Smoke Test Demo Imports

```python
from demo_app import preprocess_image, make_overlay

print("Imported demo helpers successfully:")
print("-", preprocess_image.__name__)
print("-", make_overlay.__name__)
```

## Step 4 - Launch Streamlit Demo

```python
demo_process = subprocess.Popen(
    [PYTHON, "-m", "streamlit", "run", "demo_app.py"],
    cwd=PROJECT_DIR,
)

time.sleep(3)
print("Streamlit demo should be available at http://localhost:8501")
print(f"Demo process id: {demo_process.pid}")
```

## Step 5 - Stop Streamlit Demo

```python
try:
    demo_process.terminate()
    print("Stopped Streamlit demo process.")
except NameError:
    print("No demo_process variable found in this notebook session.")
```
